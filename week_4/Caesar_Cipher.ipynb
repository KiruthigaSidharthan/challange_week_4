```python
def caesar_cipher(text, shift):
    result = []

    for x in text:
        if x.isupper():
            base = ord('A')
            shifted = (ord(x) - base + shift) % 26 + base
            result.append(chr(shifted))
        elif x.islower():
            base = ord('a')
            shifted = (ord(x) - base + shift) % 26 + base
            result.append(chr(shifted))
        else:
            result.append(x)

    return ''.join(result)


def decode_cipher(text, shift):
    return caesar_cipher(text, -shift)



shift = int(input("Enter the shift: "))
text = input("Enter the string: ")

encoded = caesar_cipher(text, shift)
decoded = decode_cipher(encoded, shift)

with open("encoded_message.txt", "w") as file:
    file.write(encoded)

print("Encoded:", encoded)
print("Decoded:", decoded)
print("Encoded message saved to encoded_message.txt")
```
