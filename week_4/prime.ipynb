```python
def is_prime():
    num = int(input("enter your input: "))
    
    if num > 1:
        for i in range(2, num):
            if num % i == 0:
                print("Not Prime")
                break
        else:
            print("Prime")
    else:
        print("Not Prime")

is_prime()
```

```python
def is_prime():
    n = int(input("enter your input: "))
    primes = []
    for num in range(2, n + 1):
        for i in range(2, int(num**0.5) + 1):
            if num % i == 0:
                break
        else:
            primes.append(num)
    return primes
print(is_prime())
```

```python
def is_prime():
    numbers = input("enter your input: ").split()
    numbers = [int(num) for num in numbers]
    
    def prime(n):
        if n < 2:
            return False    
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
    for num in numbers:
        if prime(num):
            print(f"{num} is prime")
            
        else:
            print(f"{num} is not prime")   
is_prime()
```

```python
def is_prime():
    numbers = input("enter your input: ").split()
    numbers = [int(num) for num in numbers]
    
    def prime(n):
        if n < 2:
            return False    
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True
    with open("prime_results.txt", "w") as file:    
        for num in numbers:
            if prime(num):
                result = f"{num} is prime"
                
            else:
                result = f"{num} is not prime"
            print(result)      
            file.write(result + "\n")   
    print("All results have been saved to prime_results.txt")
is_prime()
```
