```python
def find_max():
    try:

        num = input("Enter your input: ").split()
        if not num:
            print("No numbers entered! Please enter at least one number.")
            return
        nums = [int(x) for x in num]
        print(f"list is {nums}")
        current_max = nums[0]
        for x in nums:
            if x > current_max:
                current_max = x
        inx = nums.index(current_max)
        print(f"max number is {current_max} and its index is {inx}")
    except ValueError:
        print("That's not a valid number!")            
find_max()
```
