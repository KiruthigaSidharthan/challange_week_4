```python
import re
def is_palindrome():
    word  =input("Enter your input: ")
    strt =word.lower().replace(" ", "")
    str = re.sub(r'[^a-zA-Z0-9]', '', strt)
    if str == str[::-1]:
        print("Yes, it's a palindrome!")
    else:
        print("it's not a palindrome!")
is_palindrome()   
```
