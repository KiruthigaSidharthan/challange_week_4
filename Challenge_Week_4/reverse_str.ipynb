```python
def reverse_string(hello):
    print(hello[::-1])

reverse_string("Hello")
```

```python
def reverse_string():
    reverse_string = "hello"
    for str in reverse_string:
        print(reverse_string[-1],end="")
        reverse_string = reverse_string[0:-1]
reverse_string()
```

```python
import string
def reverse_string():
    reverse_string = "hello,world"
    str = ""
    for x in reverse_string:
        if x not in string.whitespace and x not in string.punctuation:
            str = x + str
    print(str)
   
reverse_string()
```
